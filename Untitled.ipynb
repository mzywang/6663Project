```python
x = []
f_x = []
grad = []
H = []
d = []

x_0 = np.array([-1.2, 1])
f_x_0, grad_0, H_0 = extended_rosenbrook(x_0)
d_0 = -grad_0

x.append(x_0)
f_x.append(f_x_0)
grad.append(grad_0)
H.append(H_0)
d.append(d_0)

print("Iteration 0")
print("x = {}, f(x) = {}".format(x[0],f_x[0]))

s = 0.5
beta = 1 / 2
sigma = 10 ** -1
epsilon = 10 ** -6
k = 1

while (f_x[-1] / f_x[0]) > epsilon:

    print("Iteration {}".format(k))
    m = 0
    x_next = x[-1] + (beta ** m) * s * d[-1]
    f_x_next, g_next, H_next = extended_rosenbrook(x_next)
    d_next = -g_next
    while (f_x[-1] - f_x_next < - sigma * (beta ** m) * s * np.inner(grad[-1], d[-1])):
        m = m + 1
        x_next = x[-1] + (beta ** m) * s * d[-1]
        f_x_next, grad_next, H_next = extended_rosenbrook(x_next)
        d_next = -grad_next
    print(x[-1])
    print(x_next)
    print(f_x[-1] - f_x_next)
    print("GRADIENT: {}".format(grad[-1]))
    print(- sigma * (beta ** m) * s * np.inner(grad[-1], d[-1]))
        
    x.append(x_next)
    f_x.append(f_x_next)
    grad.append(grad_next)
    H.append(H_next)
    d.append(d_next)
    
    print("x = {}, f(x) = {}".format(x[k],f_x[k]))
    print()
    
    k = k + 1
```
